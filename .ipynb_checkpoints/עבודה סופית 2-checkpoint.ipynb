```python
def get_part_of_finaldf(df):
    tempi = pd.DataFrame( )
    for i in range(0,5):
        temp = df[(df.index%5)==i]
        temp = temp.reset_index()
        temp = temp.drop(['index'],axis=1)
        for x in columns1:
            for y in ['Left','Right']:
                    tempi[x+' '+y+" "+str(i)] = temp[x,y]
        for x in ['Time', ' Frame ID']:
            tempi[x+' '+str(i)] = temp[x]
    return tempi
```
